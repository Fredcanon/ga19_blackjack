```python
class Card:
    valid_cards = ('A','K','Q','J','10','9','8','7','6','5','4','3','2')
    valid_suits = ('spades','hearts','diamonds','clubs')
    card_values = {'A':11,'K':10,'Q':10,'J':10,'10':10,
                   '9':9,'8':8,'7':7,'6':6,'5':5,'4':4,'3':3,'2':2}
    def __init__(self, name, suit):
        if str(name) not in Card.valid_cards:
            raise ValueError('invalid card name - must be a value 2-10 or A, K, Q, J')
        elif type(suit)!=str or suit.lower() not in Card.valid_suits:
            raise ValueError('invalid suit - must be spades, hearts, diamonds, clubs')
        else:
            self.name = str(name)
            self.suit = suit.lower()
            self.value = Card.card_values[str(name)] 
            self.description = f'{name} of {suit}'
```

```python
import random

class Deck:
    cards = ('A','K','Q','J','10','9','8','7','6','5','4','3','2')
    suits = ('spades','hearts','diamonds','clubs')

    def __init__(self):
        deck = []
        for card in Deck.cards:
            for suit in Deck.suits:
                deck.append(Card(card,suit))
        random.shuffle(deck)
        self.cards = deck
    
    def deal_card(self,display=False):
        if len(self.cards) > 0:
            card = self.cards.pop()
            if display:
                print(card.description)
            return card
        else:
            print('no more cards in deck - create a new deck')
    
    def count_cards_remaining(self):
        return len(self.cards)
```
